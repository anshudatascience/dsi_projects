## Problem Statement:

Job classification is the process to determine the relative rank of different jobs in an organization. This is important since the hierarchical position a job is classified as underlies many HR decisions. A good job classification leads to fair, equitable,senior jobs have higher requirements in terms of performance and capabilities, and enables succession planning to more senior roles.Classification involves taking the characteristics of a job and categorizing it based on the classifications, input which we will take is the job description.

Data science has been called “the sexiest job of the 21st century” — more business leaders in all industry are hiring data scientists into environments where we can be effective. 
Data Science team has been asked to classify the Data Scientist job based on the senority. The job description of 10,000 Data Scientist role has been provided to understand what could be the key terms should be mentioned on Data Scientist position. What all are the overlaps and differences in the Job description. Topic modelling has also been performed to understand the key topics in the job description.

## Pre- processing
- Pre-processing will help to transform our text.

The approach taken is as follows:

- Remove unwanted tags: As our text was collected from dataworld. Since they have some not useful   characters,  we will remove these.
- Remove non-letters: Remove special characters and numbers
- Lowercase all words and split word up
- Remove stopwords: These are common words that are not useful for text classification. 
- lemmitazing words: This will convert each word to its lemmatizing form 
-  Join words back into a string

```python
# Write a function to convert raw text to a string of meaningful words
def meaningful_text(raw_text):
    
    #REPLACE_BY_SPACE_RE = re.compile('[/(){}\[\]\|@,;]')
    #BAD_SYMBOLS_RE = re.compile('[^0-9a-z #+_]')
    
        # Remove HTML
    review_text = BeautifulSoup(raw_text).get_text()
    #review_text = REPLACE_BY_SPACE_RE.sub('  ', raw_text) # replace REPLACE_BY_SPACE_RE symbols by space in text.
    #review_text = BAD_SYMBOLS_RE.sub('  ', review_text)
    
    # Remove non-letters
    letters_only = re.sub("[^a-zA-Z]", " ", review_text)
    
    # Convert words to lower case and split each word up
    words = letters_only.lower().split()
    
    # Searching through a set is faster than searching through a list 
    # Hence, we will convert stopwords to a set
    stops = set(stopwords.words('english'))
    
    # Add certain keywords stopwords
    stops.update(['skill','year''prohibited','sex','kind',
                  'employer', 'committed','strength','diverse','workforce',
                  'read','people','saying','working','farmers','join','diverse',
                  'professionals', 'acquire', 'skills', 'apply','learned',
                  'knowledge','future','roles','offers', 'extensive',
                  'training', 'opportunities','award', 'winning', 'training',
                  'magazine', 'corporate','world', 'blackrock','investor',
                  'financial', 'fiduciary', 'investor'])
    
    # Remove stopwords
    meaningful_words = [w for w in words if w not in stops]
    
    # Stemming word
    meaningful_words = [lemmatizer.lemmatize(w) for w in meaningful_words]
   
    # Join words back into one string, with a space in between each word
    return(" ".join(meaningful_words))
```

```python
# Clean description and store it in new columns
data['clean_desc'] = data['description'].map(meaningful_text)
```

```python
# Find the no. of meaningful words in each post 
data['clean_len'] = data['clean_desc'].str.split().apply(len)
data.head(5)
```

```python
eng_stopwords = stopwords.words('english')
eng_stopwords.extend(['data','business','team','work','analytics','model','client',
                         'machine learning',"machine","learning","work","science","analytics","model",
                                  "scientist","skill","year",'analysis','ability','strong','using','new',
                     'tools','job','years','models','working','experience'])


def count_vec(df,col,n):
     
    fig, ax = plt.subplots(1, 2, figsize=(25,12))
    
    for i,col in enumerate(col):
          
        # Instantiate CountVectorizer 
        words_series = df[df['position']== col]['clean_desc']
        cvec = CountVectorizer(stop_words= eng_stopwords) 
        # Save the tokens in a dataframe
        cvec_df = pd.DataFrame(cvec.fit_transform(words_series).todense(), columns=cvec.get_feature_names())
        sum_words = cvec_df.sum(axis=0) # Sum up the no. of occurences for each word
        top_words = sum_words.sort_values(ascending=False).head(20)
    
        top_words.sort_values(ascending=True).plot(kind='barh',ax=ax[i],figsize=(12,8))
        ax[i].set_title(col, fontsize=20, fontweight='bold')
        ax[i].set_xlabel('Count', fontsize=12)
        ax[i].set_xlim(0,n) 
```

Prepare the text for LDA
For LDA, we need the number of topics, other hyperparameters. You can tweak number_topics and other hyperparameters to improve the results of Topic modeling. [**link**](https://towardsdatascience.com/evaluate-topic-model-in-python-latent-dirichlet-allocation-lda-7d57484bb5d0)

```python
# define the lemmatization function:
def lemmatization(texts, stop_words=stop_words, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):

    texts_out = []
    nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])
    nlp.max_length = 6000000
    for word in texts:
        doc = nlp(" ".join(word)) 
        # Lemmatize and only keep tokens that have certain part of speech tags
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags]) 
    # Remove stopword lemmatizatian
    texts_out = [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts_out]    
    return texts_out
```

```python
# Pre-processing text for 4 classes
data_final_0 = lemmatization(texts_0)
data_final_1 = lemmatization(texts_1)
data_final_2 = lemmatization(texts_2)
data_final_3 = lemmatization(texts_3)
```

```python
# Pre-Processing on whole corpus 
data_final = lemmatization(texts)
```

```python
def topic_model(data,n):
    # Create Dictionary
    id2word = corpora.Dictionary(data)

    # Term Document Frequency
    corpus = [id2word.doc2bow(text) for text in data]

     #Build LDA model
    lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                           id2word=id2word,
                                           num_topics=n, # number of topic
                                           random_state=100,
                                           update_every=1,
                                           chunksize=100,#controls how many documents process at one time 
                                           passes=50, #how often we train the model
                                           alpha='auto',
                                           per_word_topics=True)
      # Print the Keyword in the 5 topics
    pprint(lda_model.print_topics())
    doc_lda = lda_model[corpus]
```
